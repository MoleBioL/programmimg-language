```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://mofatarot.com/pages/taluopai-paiyi-jieshi'
```

```python
response = requests.get(url)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
import re
```

```python
pattern = re.compile(r'^https://mofatarot.com/blogs/taluo-pai-paiyi-jiexi/')
```

```python
# 找到所有的a標籤
a_tags = soup.find_all('a')

# 提取每個a標籤中的href屬性值
# 定義正則表達式來匹配開頭是 https://mofatarot.com/blogs/taluo-pai-paiyi-jiexi/ 的網址
pattern = re.compile(r'^https://mofatarot.com/blogs/taluo-pai-paiyi-jiexi/')

# 創建一個清單來存儲符合條件的 href 值
href_list = []

# 提取每個a標籤中的href屬性值，並對其應用正則表達式
for a_tag in a_tags:
    href = a_tag.get('href')
    if href and pattern.match(href):
        href_list.append(href)

# 打印清單中的 href 值
for url in href_list:
    print(url)
```

```python
# 將 href_list 轉換為集合，自動去除重複的元素，然後再轉換回列表
href_list = list(set(href_list))

# 打印清單中的 href 值
for url in href_list:
    print(url)
```

```python
# 創建 DataFrame
df = pd.DataFrame({'URL': href_list})

# 打印 DataFrame
print(df)
```
