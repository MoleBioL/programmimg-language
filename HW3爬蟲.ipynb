```python
def fetch_pages(start_page, end_page):
    conurl = 'https://read.99csw.com/book/4583'

    data = []  # 存儲每一頁的內容

    for page in range(start_page, end_page + 1):
        current_page = f'{conurl}/{page}.html'
        response = requests.get(current_page)
        soup = BeautifulSoup(response.text, 'html.parser')

        for element in soup.find_all('article', id='content'):
            content = element.text.strip()
            data.append(content)  # 將每一頁的內容添加到列表中

    df = pd.DataFrame(data, columns=['content'])  # 創建 DataFrame 並指定列名

    return df
```
